### **Read .csv from google drive**

```
college = pd.read_csv('College.csv', header=0)
college.head(n=3)
```

```
row_names = college.columns.values
print(row_names)
```

```
university_names = college.iloc[:,0]
print(university_names)
```

### **the difference between data.iloc[100,:] and data.iloc[[100]]**

```
print(college.iloc[100,:])
print(college.iloc[[100]])
```

### **reset the row names with university names**

```
#two ways
#college.set_index(university_names, inplace=True)
college.index = university_names
college = college.drop('Unnamed: 0', axis=1)
college.head()
```

###Total/Average number of applications

```
apps_total = college['Apps'].sum()
print("The total number of applications received is %d" % apps_total)
```

```
apps_avg = apps_total / college.shape[0]
print("The average number of applications received is %.2f" % apps_avg)
```

###university which has over 5000 applications

```
univ_apps_5000 = college[college['Apps']>5000]
print(univ_apps_5000.index.values)
```

```
univ_grad_5000 = univ_apps_5000['Grad.Rate'].mean(axis=0)
print("The average graduation rate for universities that received more than 5000 applications is %.2f" % univ_grad_5000)

univ_apps_4999 = college[college['Apps'] <= 5000]
univ_grad_4999 = univ_apps_4999['Grad.Rate'].mean(axis=0)
print("The average graduation rate for universities that received less than or equal to 5000 applications is %.2f" % univ_grad_4999)
```

```
high_full_grads = college[college['F.Undergrad'] == college['F.Undergrad'].max()]
print("%s has the highest number of full time undergrads" % high_full_grads.index.values[0])

low_full_grads = college[college['F.Undergrad'] == college['F.Undergrad'].min()]
print("%s has the lowest number of full time undergrads" % low_full_grads.index.values[0])
```
